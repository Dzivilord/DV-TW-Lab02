NO2, NOx và NO là các chất ô nhiễm khác cũng có tác động đáng kể nhưng tập trung vào các giờ cao điểm của giao thông, cho thấy ảnh hưởng của hoạt động di chuyển và giao thông đô thị. Các chất SO2 và CO có nồng độ thấp và ổn định, ít ảnh hưởng hơn trong ngắn hạn, nhưng vẫn cần theo dõi vì có thể tích lũy và ảnh hưởng lâu dài.

Như vậy, các biện pháp giảm thiểu ô nhiễm cần tập trung vào kiểm soát bụi mịn và Ozone vào các khung giờ nắng và giờ cao điểm để cải thiện chất lượng không khí và bảo vệ sức khỏe cộng đồng.

```python
# Giả sử dữ liệu đã được tải vào DataFrame 'df'
df = pd.read_csv("../Datasets/air_quality_shilin.csv")
# Đảm bảo rằng cột 'date' là datetime, sau đó trích xuất giờ
df['date'] = pd.to_datetime(df['date'])
df['hour'] = df['date'].dt.hour
# Tính nồng độ trung bình của từng chất ô nhiễm theo giờ
pollutants = ['so2', 'co', 'o3', 'pm10', 'pm2.5', 'no2', 'nox', 'no']
average_concentration = df.groupby('hour')[pollutants].mean()
# Vẽ heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(average_concentration.T, cmap="Reds", annot=True, fmt=".2f", cbar_kws={'label': 'Concentration'})
plt.title("Average Pollutant Concentration by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Pollutant")
plt.show()
```

- Biểu đồ line thể hiện sự thay đổi của chỉ số AQI theo từng giờ trong ngày.
- Biểu đồ bar chart giúp so sánh nồng độ của các chất ô nhiễm (O3, PM10, PM2.5) theo từng giờ. Mỗi giờ có một nhóm bar chart đại diện cho nồng độ các chất ô nhiễm tại thời điểm đó.

## 5. Kết luận
- Biểu đồ line của AQI cho thấy chỉ số chất lượng không khí thường tăng vào các giờ cao điểm trong ngày, chẳng hạn buổi sáng và buổi chiều, khi hoạt động giao thông và công nghiệp thường cao hơn.
- Nồng độ Ozone (O3) trong bar chart có xu hướng tăng vào giữa ngày, đặc biệt vào giờ trưa và đầu giờ chiều, do ánh sáng mặt trời thúc đẩy quá trình tạo ra Ozone. Điều này có thể thấy rõ khi các cột biểu diễn nồng độ O3 tăng cao trong các giờ này.
- Nồng độ PM10 và PM2.5 có xu hướng cao hơn vào các giờ đầu sáng và chiều tối, có thể liên quan đến hoạt động giao thông và công nghiệp trong các giờ này. PM2.5 thường có nồng độ cao hơn PM10, cho thấy sự ảnh hưởng của các hạt bụi siêu mịn lên chất lượng không khí.
- Từ biểu đồ, có thể kết luận rằng các giờ buổi sáng và chiều là thời điểm AQI và nồng độ các chất ô nhiễm có xu hướng cao hơn, và nồng độ Ozone đạt mức cao nhất vào buổi trưa. Các chất O3 và PM2.5 là hai chất ô nhiễm ảnh hưởng lớn đến sự thay đổi của chỉ số AQI trong ngày.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file (thay đổi đường dẫn nếu cần)
file_path = "../Datasets/air_quality_shilin.csv"
data = pd.read_csv(file_path)

# Chuyển đổi cột 'date' sang định dạng datetime
data['date'] = pd.to_datetime(data['date'])

# Trích xuất giờ trong ngày
data['hour'] = data['date'].dt.hour

# Tính trung bình nồng độ các chất ô nhiễm theo giờ
pollutants = ['so2', 'co', 'o3', 'pm10', 'pm2.5', 'no2', 'nox', 'no']
mean_pollutants_by_hour = data.groupby('hour')[pollutants].mean()

# Tạo heatmap với màu từ trắng sang đỏ
plt.figure(figsize=(12, 6))
sns.heatmap(mean_pollutants_by_hour, annot=True, fmt=".2f", cmap="Reds", cbar_kws={'label': 'Nồng độ'})
plt.title('Trung bình nồng độ chất ô nhiễm theo giờ')
plt.xlabel('Chất ô nhiễm')
plt.ylabel('Giờ trong ngày')
plt.show()
```

   - **CO** và **SO2** có ảnh hưởng nhỏ hơn, với sự biến động không quá lớn và không có sự tương quan rõ rệt với AQI.

#### **Kết luận cuối cùng**:
- **PM2.5** và **PM10** có ảnh hưởng mạnh đến AQI và cần được kiểm soát chặt chẽ trong những giờ cao điểm (buổi sáng).
- **Ozone (O3)** cần được chú ý vào buổi chiều do sự gia tăng nồng độ.
- **SO2** và **CO** ít ảnh hưởng đến chất lượng không khí, nhưng vẫn cần có biện pháp kiểm soát liên tục.
- Việc giảm phát thải từ giao thông, cải thiện điều kiện giao thông và giảm lượng các chất ô nhiễm vào các giờ cao điểm có thể giúp cải thiện chất lượng không khí ở Shilin, Đài Loan.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu từ file
file_path = "../Datasets/air_quality_shilin.csv"
data = pd.read_csv(file_path)

# Chuyển đổi cột 'date' sang định dạng datetime
data['date'] = pd.to_datetime(data['date'])

# Trích xuất giờ trong ngày
data['hour'] = data['date'].dt.hour

# Tính trung bình nồng độ AQI và các chất ô nhiễm theo giờ
pollutants = ['so2', 'co', 'o3', 'pm10', 'pm2.5']
grouped_data = data.groupby('hour')[['aqi'] + pollutants].mean()

# Vẽ biểu đồ
fig, ax1 = plt.subplots(figsize=(14, 8))

# Trục Y thứ nhất: Biểu đồ đường cho AQI
ax1.plot(grouped_data.index, grouped_data['aqi'], color='blue', label='AQI', linewidth=2)
ax1.set_ylabel('AQI', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xlabel('Hour of Day', fontsize=12)

# Trục Y thứ hai: Biểu đồ cột cho các chất ô nhiễm
ax2 = ax1.twinx()
width = 0.15  # Độ rộng của các cột
x = grouped_data.index

# Tạo cột cho từng chất ô nhiễm
colors = ['red', 'orange', 'green', 'purple', 'brown']
for i, pollutant in enumerate(pollutants):
    ax2.bar(x + i * width - (len(pollutants) / 2) * width, grouped_data[pollutant],
            width=width, label=pollutant, color=colors[i])

ax2.set_ylabel('Nồng độ chất ô nhiễm', fontsize=12)
ax2.tick_params(axis='y')
ax2.set_ylim(0, grouped_data[pollutants].max().max() * 1.2)  # Điều chỉnh giới hạn trục Y

# Thêm chú thích và tiêu đề
fig.legend(loc='upper right', bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)
plt.title('Chỉ số AQI và nồng độ chất ô nhiễm theo giờ', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Biểu đồ violin là lựa chọn phù hợp để phân tích sự thay đổi nồng độ của các chất ô nhiễm theo thời gian. Nó cung cấp cái nhìn trực quan về sự phân bố, độ tập trung và biến động của dữ liệu.

#### **Kết luận**:
1. **PM2.5**:
   - Nồng độ **PM2.5** thường tăng cao vào khoảng **6-8 giờ sáng**, có thể liên quan đến hoạt động giao thông giờ cao điểm.
   - Một số giờ có phân phối rộng, cho thấy nồng độ PM2.5 thay đổi đáng kể trong các ngày khác nhau.

2. **NO (Nitric Oxide)**:
   - **NO** có xu hướng nồng độ cao vào buổi sáng sớm (6-8 giờ) và giảm dần trong ngày.
   - Phân bố của NO khá chặt chẽ vào sáng sớm, cho thấy đây là thời điểm nồng độ ổn định nhất.

3. **NO2 (Nitrogen Dioxide)**:
   - Nồng độ **NO2** đạt đỉnh vào sáng sớm (khoảng 6-8 giờ) và tối muộn (khoảng 18-20 giờ). Điều này có thể liên quan đến hoạt động giao thông và thời điểm đốt nhiên liệu.
   - Biểu đồ NO2 cho thấy có sự biến động lớn vào một số giờ, với nhiều giá trị ngoại lai.

4. **So sánh các chất**:
   - Cả **PM2.5**, **NO**, và **NO2** đều đạt nồng độ cao vào giờ cao điểm buổi sáng (6-8 giờ), thể hiện rõ ràng tác động của các hoạt động giao thông.
   - Nồng độ NO và NO2 có xu hướng giảm dần vào giữa ngày, trong khi PM2.5 duy trì ở mức trung bình cao hơn.

#### **Kết luận cuối cùng**:
- **PM2.5** là chất ô nhiễm có phân bố rộng nhất, thể hiện mức độ ảnh hưởng lớn và sự thay đổi đáng kể giữa các giờ trong ngày.
- **NO** và **NO2** cũng ảnh hưởng mạnh, đặc biệt vào giờ cao điểm buổi sáng và tối muộn, cho thấy mối liên hệ trực tiếp với hoạt động giao thông và phát thải nhiên liệu.
- Các biện pháp kiểm soát cần tập trung vào việc giảm phát thải từ giao thông trong khung giờ cao điểm để cải thiện chất lượng không khí tại Shilin, Đài Loan.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file (thay đổi đường dẫn nếu cần)
file_path = "../Datasets/air_quality_shilin.csv"
data = pd.read_csv(file_path)

# Chuyển đổi cột 'date' sang định dạng datetime
data['date'] = pd.to_datetime(data['date'])

# Trích xuất giờ trong ngày
data['hour'] = data['date'].dt.hour

# Danh sách các chất ô nhiễm cần vẽ
pollutants = ['pm2.5', 'no', 'no2']

# Tạo các subplot
fig, axes = plt.subplots(3, 1, figsize=(18, 6), sharey=True)

# Lặp qua các chất ô nhiễm và vẽ từng violin plot
for i, pollutant in enumerate(pollutants):
    sns.violinplot(x='hour', y=pollutant, data=data, palette='coolwarm', inner='box', ax=axes[i])
    axes[i].set_title(f'Biểu đồ Violin của {pollutant.upper()} theo giờ', fontsize=14)
    axes[i].set_xlabel('Giờ', fontsize=12)
    #axes[i].set_ylabel(f'{pollutant.upper()} Concentration' if i == 0 else '', fontsize=12)
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Tinh chỉnh khoảng cách giữa các subplot
plt.tight_layout()
plt.show()
```

# AQI PHỤ THUỘC VÀO NHỮNG CHẤT Ô NHIỄM VÀ YẾU TỐ KHÍ TƯỢNG NÀO NHIỀU NHẤT Ở SHILIN, ĐÀI LOAN TRONG GIAI ĐOẠN 2016 - 2024?

## 1. Các thuộc tính sử dụng
- `aqi`: Chỉ số chất lượng không khí.
- Các chất ô nhiễm: `pm2.5`, `pm10`, `so2`, `no2`, `co`, `o3`.
- Các yếu tố khí tượng: `windspeed` (tốc độ gió), `winddirec` (hướng gió).

## 2. Loại biểu đồ
- Biểu đồ radar (Radar Chart).

## 3. Ý nghĩa của biểu đồ
Biểu đồ radar biểu diễn mức độ tương quan giữa **AQI** và các yếu tố khác (bao gồm chất ô nhiễm và yếu tố khí tượng). Các trục thể hiện các yếu tố, và giá trị trên mỗi trục chỉ ra mức độ phụ thuộc của AQI vào yếu tố đó. 

- Tương quan dương (giá trị > 0): Yếu tố làm **tăng** AQI.
- Tương quan âm (giá trị < 0): Yếu tố làm **giảm** AQI.

## 4. Nguyên nhân sử dụng
Biểu đồ radar cho phép trực quan hóa và so sánh mức độ ảnh hưởng của nhiều yếu tố đồng thời. Việc sử dụng biểu đồ radar giúp nhận diện yếu tố nào có mối tương quan chặt chẽ nhất (cả tích cực và tiêu cực) với AQI.

## 5. Kết luận
1. **PM2.5 và PM10 là hai yếu tố ảnh hưởng mạnh nhất đến AQI:**
   - PM2.5 có giá trị tương quan dương cao, cho thấy nó là yếu tố chính làm tăng chỉ số AQI. Đây là hạt mịn có khả năng tồn tại lâu trong không khí và gây ô nhiễm nghiêm trọng.
   - PM10 cũng có tương quan dương đáng kể, mặc dù thấp hơn so với PM2.5, cho thấy nó cũng đóng vai trò lớn trong việc ảnh hưởng đến AQI.

2. **Ozone (O3) cũng có mức độ ảnh hưởng đáng kể:**
   - Giá trị tương quan dương của ozone cho thấy nó có tác động đến AQI, đặc biệt trong điều kiện thời tiết nắng nóng khi O3 tăng cao.

3. **Yếu tố khí tượng - Windspeed và Winddirec:**
   - **Windspeed** (tốc độ gió) có tương quan âm nhẹ với AQI, nghĩa là gió mạnh giúp khuếch tán các chất ô nhiễm và làm giảm chỉ số AQI.
   - **Winddirec** (hướng gió) không có tương quan đáng kể, cho thấy không ảnh hưởng nhiều đến AQI.

4. **Các chất ô nhiễm khác như SO2, NO2 và CO có ảnh hưởng thấp hơn:**
   - Tương quan thấp hoặc gần bằng 0 giữa AQI và các chất này cho thấy chúng không phải là yếu tố chính ảnh hưởng đến chất lượng không khí.

## **Kết luận cuối cùng**:
- **PM2.5** là yếu tố ảnh hưởng mạnh nhất đến AQI, tiếp theo là **PM10** và **Ozone (O3)**.  
- Yếu tố khí tượng như **windspeed** có vai trò giảm nhẹ AQI nhưng không đủ lớn để tạo ra thay đổi đáng kể.  
- Các chất SO2, NO2 và CO có ảnh hưởng rất thấp, không đáng kể.  

Các biện pháp kiểm soát cần tập trung vào việc giảm lượng PM2.5 và PM10 để cải thiện chất lượng không khí tại Shilin, Đài Loan.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file (thay đổi đường dẫn nếu cần)
file_path = "../Datasets/air_quality_shilin.csv"
data = pd.read_csv(file_path)

# Danh sách các yếu tố ảnh hưởng đến AQI
factors = ['so2', 'co', 'o3', 'pm10', 'pm2.5', 'no2', 'windspeed', 'winddirec']

# Tính hệ số tương quan giữa AQI và các yếu tố
correlations = data[factors + ['aqi']].corr()['aqi'][:-1]

# Tạo dữ liệu cho biểu đồ radar
labels = correlations.index
values = correlations.values

# Thêm điểm đầu để khép kín biểu đồ radar
labels = np.concatenate([labels, [labels[0]]])
values = np.concatenate([values, [values[0]]])

# Vẽ biểu đồ radar
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=True)

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

# Biểu đồ radar
ax.plot(angles, values, linewidth=2, linestyle='solid', label='Tương quan với AQI')
ax.fill(angles, values, color='yellow', alpha=0.3)

# Cài đặt nhãn và tiêu đề
ax.set_xticks(angles)
ax.set_xticklabels(labels, fontsize=12)
ax.set_title('Sự phụ thuộc của Chỉ số Chất lượng Không khí (AQI) vào các chất ô nhiễm và yếu tố khí tượng', fontsize=14, pad=20)

# Điều chỉnh trục Y
ax.set_rlabel_position(30)
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.set_yticklabels([-1, -0.5, 0, 0.5, 1], fontsize=10)
ax.set_ylim(-1, 1)

# Hiển thị
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.tight_layout()
plt.show()
```
